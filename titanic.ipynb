```python
import pandas as pd

df = pd.read_csv('/content/train.csv')
df.head(20)
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df['Embarked'].mode()[0] # valor mais frequente
```

```python
df.isnull().sum() / df.shape[0]
```

```python
df.drop(['Name', 'Ticket', 'PassengerId'], axis=1, inplace=True)
```

```python
df['Cabin'].value_counts()
```

```python
df.drop('Cabin', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
sex_encoded = pd.get_dummies(df['Sex'])

embarked_encoded = pd.get_dummies(df['Embarked'])

df.drop('Sex', inplace=True, axis=1)
df.drop('Embarked', inplace=True, axis=1)

concatenado = pd.concat([df, sex_encoded, embarked_encoded], axis=1)
```

```python
concatenado.head()
```

```python
y = concatenado['Survived']
x = concatenado.drop('Survived', axis=1)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score

modelo1 = GradientBoostingClassifier(n_estimators=300, learning_rate=0.09)
kfold = StratifiedKFold(n_splits=5)

resultado = cross_val_score(modelo1, x, y, cv=kfold, n_jobs=-1)
print(resultado.mean())
```

## Melhores parmetros

```python
import numpy as np
from sklearn.model_selection import GridSearchCV

valores_grid = {
    'learning_rate': [0.1, 0.09, 0.08, 0.07, 0.06, 0.05, 0.04, 0.03, 0.02, 0.01]
}

modelo2 = GradientBoostingClassifier(n_estimators=300)

gridGB = GridSearchCV(estimator=modelo2, param_grid=valores_grid, cv=5, n_jobs=-1)
gridGB.fit(x, y)

print(gridGB.best_estimator_.learning_rate)
print(gridGB.best_score_)
```
